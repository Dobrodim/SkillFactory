# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


```python
answers['27'] = 'Daniel Radcliffe & Rupert Grint'
```

```python
data27 = data.copy()
data27.cast = data27.cast.str.split('|')
cast_pairs = []
for actors in data27.cast:
    comb = combinations(actors, 2)
    pairs = [', '.join(i) for i in comb]
    cast_pairs.append(pairs)   
cast_pairs = list(chain.from_iterable(cast_pairs))
total_pairs = Counter([item for item in sorted(cast_pairs, key=lambda x: 
                cast_pairs.count(x), reverse=True)])
total_pairs
```
